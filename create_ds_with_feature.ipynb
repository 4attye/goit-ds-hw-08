```python
from pathlib import Path
import pandas as pd
import  numpy as np

folders = ['idle', 'running', 'stairs', 'walking']
final_dataset = []

for folder in folders:
    data = Path(f'data/{folder}')
    files = [f for f in data.glob('*.csv')]
    column_activity = f'{folder}'
    merged_file = []
    for file in files:
        df = pd.read_csv(file)
        for data in df.columns[:3]:
            df[f'{data[-1]}_mean']= df[data].mean()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_variance']= df[data].var()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_std']= df[data].std()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_median']= df[data].median()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_min']= df[data].min()   
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_max']= df[data].max()  
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_range']= df[data].max() - df[data].min()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_rms']= np.sqrt(np.mean(np.square(df[data])))  
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_signal_area']= df[data].sum()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_min_index']= df[data].argmin()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_max_index']= df[data].argmax()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_power']= (df[data]**2).mean()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_energy']= (df[data]**2).sum()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_skewness']= (df[data]**2).skew()
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_kurtosis']= (df[data]**2).kurtosis()
            
        for data in df.columns[:3]:
            Q1 = df[data].quantile(.25)
            Q3 = df[data].quantile(.75)
            df[f'{data[-1]}_IQR']= Q3 - Q1
            
        for data in df.columns[:3]:
            df[f'{data[-1]}_mean_absolute_deviation']= (df[data] - df[data].mean()).abs().mean()
            
        for data in df.columns[:3]:
            hist, _ = np.histogram(df[data], bins=10, density=True)
            hist = hist[hist > 0]
            df[f'{data[-1]}_entropy']= -np.sum(hist * np.log2(hist))
      
        df.insert(0, 'activity', f'{column_activity}')
        df_drop_columns = df.drop(columns=['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z'])
        df_drop_duplicates = df_drop_columns.drop_duplicates()
        
        merged_file.append(df_drop_duplicates)
    df_features = pd.concat(merged_file, ignore_index=True)
    final_dataset.append(df_features)
    
ds_feature = pd.concat(final_dataset, ignore_index=True)
ds_feature
```

```python
ds_feature.to_csv('ds_feature.csv')
```
